## Import

```python
import tensorflow as tf
from tensorflow import keras
from keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
import numpy as np
```

## Data

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
x_train[0]
```

```python
x_train.shape, x_test.shape
```

```python
idx = 10
plt.imshow(x_train[idx], cmap='gray')
print(y_train[idx])
```

## Preprocessing

```python
np.min(x_train), np.max(x_train)
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
np.min(x_train), np.max(x_train)
```

## Model

## Dense Layer - General Deep Learning model

```python
model = keras.Sequential()

model.add(keras.layers.Input(shape=(28,28)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(256, activation='relu'))
model.add(keras.layers.Dense(256, activation='relu'))
model.add(keras.layers.Dense(256, activation='relu'))
model.add(keras.layers.Dense(256, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
hist = model.fit(x_train, y_train, batch_size=512, validation_data=(x_test, y_test), epochs=500)
```

##CNN Model

```python
model_cnn = keras.Sequential()

model_cnn.add(keras.layers.Conv2D(64, (5,5), input_shape=(28,28,1), padding='same'))
model_cnn.add(keras.layers.BatchNormalization())
model_cnn.add(keras.layers.Activation('relu'))
model_cnn.add(keras.layers.MaxPool2D())

model_cnn.add(keras.layers.Conv2D(128, (3,3)))
model_cnn.add(keras.layers.BatchNormalization())
model_cnn.add(keras.layers.Activation('relu'))
model_cnn.add(keras.layers.AveragePooling2D(pool_size=(2,2)))

model_cnn.add(keras.layers.Conv2D(128, (3,3)))
model_cnn.add(keras.layers.BatchNormalization())
model_cnn.add(keras.layers.Activation('relu'))
model_cnn.add(keras.layers.MaxPool2D())

model_cnn.add(keras.layers.Flatten())
model_cnn.add(keras.layers.Dense(128, activation='relu'))
model_cnn.add(keras.layers.Dropout(0.5))
model_cnn.add(keras.layers.Dense(128, activation='relu'))
model_cnn.add(keras.layers.Dropout(0.5))
model_cnn.add(keras.layers.Dense(10, activation='softmax'))
```

```python
model_cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model_cnn.summary()
```

```python
hist_cnn = model_cnn.fit(x_train, y_train, batch_size=512, validation_data=(x_test, y_test), epochs=500)
```

## Evaluation

```python
plt.plot(hist.history['accuracy'], color='green')
plt.plot(hist_cnn.history['accuracy'], color='blue')
```

```python
plt.plot(hist.history['val_accuracy'], color='green')
plt.plot(hist_cnn.history['val_accuracy'], color='blue')
```

```python
plt.plot(hist_cnn.history['accuracy'], color='blue')
plt.plot(hist_cnn.history['val_accuracy'], color='red')
```
