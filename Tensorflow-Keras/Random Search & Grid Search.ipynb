```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=1000, n_features=20, n_informative=10, n_redundant=5, n_classes=2, random_state=0)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape
```

# Model Definition

```python
from tensorflow import keras
import tensorflow as tf
```

```python
def create_model(hidden_units=16, optimizer='adam'):
    model = keras.Sequential()
    model.add(keras.layers.Dense(units=hidden_units, input_shape=(20,), activation='relu'))
    model.add(keras.layers.Dense(units=1, activation='sigmoid'))

    model.compile(optimizer=optimizer, loss=tf.losses.binary_crossentropy, metrics=['accuracy'])

    return model
```

# Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
# pip install -U scikit-learn==1.4.2 scikeras==0.13.0
```

```python
from scikeras.wrappers import KerasClassifier
class FixedKerasClassifier(KerasClassifier):
    def __sklearn_tags__(self):
        return super()._more_tags()
```

```python
model = KerasClassifier(model=create_model, loss='binary_crossentropy', epochs=20, batch_size=32, verbose=0)
```

```python
params = {
    'model__hidden_units':[16, 32, 64],
    'model__optimizer':['adam', 'rmsprop']
}
```

```python
grid = GridSearchCV(estimator=model, param_grid=params, cv=3)
grid_result = grid.fit(X_train, y_train)
```

```python
best_param = grid_result.best_params_
best_param
```

```python
best_score = grid_result.best_score_
best_score
```

# Random Search

```python
from sklearn.model_selection import RandomizedSearchCV

rs = RandomizedSearchCV(estimator=model, param_distributions=params, cv=3, n_iter=4)
rs_result = rs.fit(X_train, y_train)
```

```python
rs_best_params = rs_result.best_params_
rs_best_params
```

```python
rs_best_score = rs_result.best_score_
rs_best_score * 100
```
